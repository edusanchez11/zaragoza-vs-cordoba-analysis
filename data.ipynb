```python
# Obtener todas las estadísticas y filtrar Zaragoza vs Córdoba
import pandas as pd

df_total = fbref.get_all_teams_season_stats('La Liga 2', '2025-2026')

# Filtrar equipos
equipos_filtrados = []
for idx, row in df_total.iterrows():
    squad = str(row['Squad']).lower()
    if 'zaragoza' in squad or 'córdoba' in squad or 'cordoba' in squad:
        equipos_filtrados.append(idx)

# Mostrar resultado
if equipos_filtrados:
    total_zaragoza_cordoba_filtrado = df_total.loc[equipos_filtrados]
    pd.set_option('display.max_columns', None)
    total_zaragoza_cordoba_filtrado
else:
    print("Equipos no encontrados")
```

```python
# Mostrar el DataFrame completo filtrado
total_zaragoza_cordoba_filtrado
```

```python
# guardar el dataframe filtrado
total_zaragoza_cordoba_filtrado.to_csv('total_zaragoza_vs_cordoba.csv', index=False)
```
